```python
def dataset_copy_tb_images(lst_tbud_ids, path_images, path_tbud_txts, path_target):
    """
    
    """
    def zeroPad(idOf, length=4):
        i = 0
        l = length - len(idOf)
        while i < l:
            idOf = '0' + idOf
            i +=1
        return idOf

    for tbud_id in lst_tbud_ids:
    
        dir_imgs = os.path.join(path_images, tbud_id + '_conv/images/')
        dir_txts = os.path.join(path_tbud_txts, tbud_id + '_Bud_Info_txt/')

        _, _, files_imgs = next(os.walk(dir_imgs))
        _, _, files_txts = next(os.walk(dir_txts))

        files_imgs = sorted(files_imgs)
        files_txts = sorted(files_txts)
        
        if not os.path.exists(path_target):
            os.makedirs(path_target)

        start_time = time.time()
        print(f'[DEBUG] Started Tbud {tbud_id} Included Image Tile Copy...')

        for i in range(len(files_txts)):

            if files_txts[i].split('-')[0] == tbud_id:

                col = files_txts[i].split('-')[1]
                row = files_txts[i].split('-')[2]

                fname_img = dir_imgs + zeroPad(col) + '-' + zeroPad(row) + '.jpg'
                fname_written = 'orj-' + str(tbud_id) + '-' + col + '-' + row + '.jpg'
                dir_write = os.path.join(path_target, 'orj',str(tbud_id))
                
                if not os.path.exists(dir_write): os.makedirs(dir_write)
                copyfile(fname_img, os.path.join(dir_write, fname_written))

        print(f'[DEBUG] Finished Tbud {tbud_id} Image Tile Copy to Destination {path_target}! \nTime: {(time.time() - start_time)}')
```

# Resize and Save

```python
def dataset_resize_orj_images(lst_tbud_ids, path_images, path_target, size_target=512):
    """
    
    """
    def remove_zero(ids):
        if ids == '0000': return 0
        else: return ids.lstrip('0')

    for tbud_id in lst_tbud_ids:
    
        dir_imgs = os.path.join(path_images, tbud_id + '_conv/images/')
        _, _, files_imgs = next(os.walk(dir_imgs))
        files_imgs = sorted(files_imgs)
        
        if not os.path.exists(path_target):
            os.makedirs(path_target)

        start_time = time.time()
        print(f'[DEBUG] Started Tbud {tbud_id} Included Image Tile Copy...')

        for i, file in enumerate(files_imgs):
            file_split = file.split('.')[0]
            col = str(remove_zero(file_split.split('-')[0]))
            row = str(remove_zero(file_split.split('-')[1]))

            fname_img = dir_imgs + file
            print(fname_img)
            img = cv2.imread(fname_img, cv2.IMREAD_COLOR)
            img_resized = cv2.resize(img, (512, 512), interpolation=cv2.INTER_AREA)

            fname_written = 'orj-' + str(size_target) + '-' + str(tbud_id) + '-' + col + '-' + row + '.jpg'
            dir_write = os.path.join(path_target, 'orj-' + str(size_target), str(tbud_id))

            if not os.path.exists(dir_write): 
                os.makedirs(dir_write)
            print(os.path.join(dir_write, fname_written))
            cv2.imwrite(os.path.join(dir_write, fname_written), img_resized)
                
        print(f'[DEBUG] Finished Tbud {tbud_id} Image Tile Resize and Write to Destination {path_target}! \nTime: {(time.time() - start_time)}')
```
